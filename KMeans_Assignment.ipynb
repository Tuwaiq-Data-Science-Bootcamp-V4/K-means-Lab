# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.scatterplot(data = df, x= "Room.Board" , y = "Grad.Rate",hue="Private" ,color= "Private" , alpha = 0.9)
plt.legend(title='Private',loc='center right')
plt.show()
```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")

sns.scatterplot(data = df, x= "Outstate" , y = "F.Undergrad",hue="Private" ,color= "Private" , alpha = 0.3)
plt.legend(title='Private')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')

fig = sns.FacetGrid(data = df,hue="Private", height=6 , aspect=2 )
fig.map(sns.histplot, "Outstate" ,  multiple="stack",  alpha=0.3)
plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')

fig2= sns.FacetGrid(data = df,hue="Private", height=6 , aspect=2 )
fig2.map(sns.histplot, "Grad.Rate" ,  alpha=0.3)
plt.show()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
data = df.drop('Private',axis=1)
```

```python
y_kmeans = kmeans.fit_predict(data)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df["Cluster"]= np.where(df["Private"] == "Yes" ,1 ,0)
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(confusion_matrix(df['Cluster'], y_kmeans))
```

```python
print(classification_report(df['Cluster'], y_kmeans))
```
