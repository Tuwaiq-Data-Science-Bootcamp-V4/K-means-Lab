## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
X = df.drop(columns=['Private'])
X  = np.array(X)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
#create cluster with values 0
df['Cluster']=0
#if school is private -> change value to 1
df['Cluster'] = np.where(df['Private'] == 'Yes', 1 ,df['Cluster'])
#check data
df
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import classification_report, confusion_matrix


print(confusion_matrix(df['Cluster'],y_kmeans))
```

```python
print(classification_report(df['Cluster'],y_kmeans))
```
